To what extent does the presence of a professional sports stadium act as a localized economic catalyst compared to broader metropolitan trends? Specifically, does the percentage growth in median property values (wealth), total business licenses (commercial activity), and employment rates (opportunity) of neighbourhoods within a 2 mile radius of a stadium significantly outpace the growth of the surrounding neighbourhoods within the same city? 

## Background and Prior Work

Professional sports stadiums are often significantly large venues that host often profitable sporting events for the NFL, NBA, NHL, and other sporting leagues. Stadiums are also culturally significant as events like the Super Bowl, hosted in the Caesars Superdome, New Orleans, Louisiana in 2025, had around 120 million viewers and around 76,000 filled seats.It can be assumed that stadiums  stimulate economic activity around the local area as businesses take advantage of the large foot traffic stadiums create, creating more jobs, increasing spending, tourism, and property value <sup>[2](#https://www.researchgate.net/publication/46552225_How_Does_a_New_Sports_Stadium_Affect_Housing_Values_The_Case_of_Fedex_Field)</sup>. 


However, studies and articles conclude that the construction of new stadiums yields negligible improvements to local employment<sup>[3](#https://doi.org/10.1002/pam.22534)</sup>, and may take spending away from local businesses as people spend it on stadium related costs instead. Additionally, stadium construction is often funded by public subsidies that can be used on other local construction projects or programs. While stadiums may generate a localized economic surge within their immediate vicinity<sup>[1](#https://www.aimspress.com/article/doi/10.3934/NAR.2024024?viewType=HTML)</sup> and foster civic identity and pride, analysts argue their regional economic impact is limited.


For this project, we are observing how professional sports stadiums affect the local economy. In a 2024 research paper, Antolini Valerio analyzes trends in GDP per capita in Turin, Italy around the construction of a new and technologically impressive NFL stadium called a next-generation stadium. The writer found a causal relationship between an increase in GDP per capita with the new stadium compared to other regions of Italy that showed comparatively lower growths in GDP per capita1. The paper provides some insights on how new and expensive stadiums can change local economic production. In a study by Charles Tu, the construction of a stadium causes homes to be sold at a lower price, but over longer periods of time land value increases. The study shows that there are changes to local property values which will be examined and compared in our own data. The paper by Bradbury and co, details the large public costs of newly constructed professional sports stadiums compared to how little economic improvement they provide. This review on the stadiums and their impacts show the confounding ethical, public, and economic variables that we need to interpret to understand how professional sports stadiums impact localized economies.

# 1. Industry

This variable describes the sector in which employed individuals work. Examples include:

1. Agriculture, forestry, fishing, and mining
2. Construction
3. Manufacturing
4. Retail trade
5. Professional, scientific, and management services
6. Education and healthcare services
7. Public administration

Industry is reported as a count of workers or as a percentage of the employed population in each sector.

This metric helps describe the economic structure of a subdivision. For example:
- A high percentage in agriculture may indicate a rural economy.
- A high percentage in professional services may indicate an urban or higher-income area.

# 2. Employment Status

This variable categorizes the working-age population into groups such as:
- Employed
- Unemployed (actively seeking work)
- Not in labor force

Values are reported as counts or percentages (%) of the civilian labor force.

The unemployment rate is particularly important:
- In stable economic conditions, unemployment is often around 3–6%.
- Rates significantly above 8–10% may indicate local economic distress.

Labor force participation rate is also included, which measures the percentage of the population that is either working or actively seeking work.

# 3. Commuting to Work

This category describes how people travel to work and includes:
- Car, truck, or van
- Public transportation
- Walked
- Worked from home
- Other means

It also includes average commute time (minutes).

Units are:
- Percent of workers by transportation mode (%)
- Mean commute time in minutes
- This variable gives insight into infrastructure, urbanization, and economic integration with nearby regions.

# 4.Occupation

Occupation refers to the type of job performed, such as:
- Management, business, science, and arts occupations
- Service occupations
- Sales and office occupations
- Natural resources, construction, and maintenance occupations
- Production, transportation, and material moving occupations

Values are usually percentages of employed persons.

Occupation differs from industry:
1. Industry describes what sector the employer is in.
2. Occupation describes what type of job the person performs.

For example, an accountant working for a hospital is:
- Industry: Healthcare
- Occupation: Business/Financial operations 

5. Income and Benefits
This category typically includes:
- Median household income (USD per year)
- Per capita income (USD per year)
- Mean earnings (USD per year)
- Receipt of public assistance (%)
- Social Security income (%)

Income values are measured in U.S. dollars per year, and median household income is particularly important:
- National median income in recent years is around $70,000–$75,000.
- Values significantly below $40,000 may indicate economic hardship.
- Values above $100,000 often reflect affluent areas.

Because income distributions are often skewed, the median is usually preferred over the mean.

6. Health Insurance Coverage

This variable measures:
- Percent insured
- Percent uninsured

Units are percentages (%).

The uninsured rate is important:
- National uninsured rates are often around 8–10%.
- Rates above 15–20% may indicate limited access to healthcare.

This metric is closely tied to employment patterns, income, and state-level Medicaid expansion policies.

7. Percentage of Families and People Below the Poverty Level

This variable measures the percentage of:
- Individuals below poverty
- Families below poverty
- Units: percentage (%)

The federal poverty threshold is defined annually by the federal government and depends on family size and composition.

Interpretation:
- U.S. poverty rate is typically around 10–15%.
- Rates above 20% may indicate significant economic vulnerability.
- Extremely low poverty rates (<5%) often occur in affluent suburbs.

This metric is one of the most important indicators of economic distress.

# Concerns with the Dataset

## 1. Sampling Error (ACS-Based Estimates)
If this dataset comes from the American Community Survey (ACS), it is based on a sample rather than a full census. Smaller subdivisions often have:
- Larger margins of error
- Less stable estimates
- Small population areas may show exaggerated variation.

## 2. Suppression or Missing Data
Some Zip Codes may have missing values due to:
- Small population size
- Confidentiality protection
- Insufficient survey responses

## 3. Self-Reported Data
Income, employment status, and insurance coverage are self-reported, which introduces potential reporting bias.

## 4. Comparability Across Zip Codes

Zip Codes do not determine the number of people in that area, and some zip codes can be inflated and have many more people, or a skewed wealth distribution.
