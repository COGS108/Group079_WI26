## Background and Prior Work

Professional sports stadiums are often significantly large venues that host often profitable sporting events for the NFL, NBA, NHL, and other sporting leagues. Stadiums are also culturally significant as events like the Super Bowl, hosted in the Caesars Superdome, New Orleans, Louisiana in 2025, had around 120 million viewers and around 76,000 filled seats.It can be assumed that stadiums  stimulate economic activity around the local area as businesses take advantage of the large foot traffic stadiums create, creating more jobs, increasing spending, tourism, and property value <sup>[2](#https://www.researchgate.net/publication/46552225_How_Does_a_New_Sports_Stadium_Affect_Housing_Values_The_Case_of_Fedex_Field)</sup>. 


However, studies and articles conclude that the construction of new stadiums yields negligible improvements to local employment<sup>[3](#https://doi.org/10.1002/pam.22534)</sup>, and may take spending away from local businesses as people spend it on stadium related costs instead. Additionally, stadium construction is often funded by public subsidies that can be used on other local construction projects or programs. While stadiums may generate a localized economic surge within their immediate vicinity<sup>[1](#https://www.aimspress.com/article/doi/10.3934/NAR.2024024?viewType=HTML)</sup> and foster civic identity and pride, analysts argue their regional economic impact is limited.


For this project, we are observing how professional sports stadiums affect the local economy. In a 2024 research paper, Antolini Valerio analyzes trends in GDP per capita in Turin, Italy around the construction of a new and technologically impressive NFL stadium called a next-generation stadium. The writer found a causal relationship between an increase in GDP per capita with the new stadium compared to other regions of Italy that showed comparatively lower growths in GDP per capita1. The paper provides some insights on how new and expensive stadiums can change local economic production. In a study by Charles Tu, the construction of a stadium causes homes to be sold at a lower price, but over longer periods of time land value increases. ### Dataset #2 Paragraphs

The most important metrics in the ACS Demographic and Housing Estimates dataset include total population, race and ethnicity composition, and total housing units. Total population is measured as an estimated count of individuals living in a given geographic area (such as a census tract, city, or state). For example, a census tract might have an estimated population of 4,200 people. These counts allow us to compare neighborhood size and calculate per-capita measures when analyzing economic outcomes.

Race and ethnicity variables are also reported as counts and percentages, showing the share of residents identifying as White, Black or African American, Asian, Hispanic or Latino, and other categories. Changes in these percentages over time can signal demographic shifts such as migration, redevelopment, or possible displacement.

Total housing units is measued as a count of housing units in a specified area.

There are several major concerns with this dataset. First, because the ACS is survey based, estimates are subject to sampling error and potential nonresponse bias. Certain populations, such as undocumented residents or lower-income households may be undercounted. Second, smaller geographic units like census tracts often have larger margins of error, which can reduce confidence in precise comparisons. Finally, the dataset provides aggregated summary statistics rather than individual level data, meaning we cannot directly observe displacement or track specific residents over time. 

Additionaly, demographic changes observed near stadiums may reflect broader citywide economic trends rather than the direct impact of the stadium itself, making careful statistical design necessary to avoid misleading conclusions that may be caused by other factors.

Note: The loading and parsing shown here is done on all ZIP Code tabulation areas in the united states.

```python
import pandas as pd
from pathlib import Path

raw_data_path = "data/00-raw/ACSDP5Y2024.DP05-Data.csv"
metadata_path = "data/00-raw/ACSDP5Y2024.DP05-Column-Metadata.csv"

df_raw = pd.read_csv(raw_data_path, encoding="utf-8-sig", low_memory=False)

cols_to_keep = {
    "GEO_ID": "geo_id",
    "NAME": "area_name",
    "DP05_0001E": "total_population",
    "DP05_0037E": "white_alone",
    "DP05_0045E": "black_alone",
    "DP05_0061E": "asian_alone",
    "DP05_0090E": "hispanic_or_latino",
    "DP05_0105E": "total_housing_units",
    "DP05_0105M": "moe_total_housing_units"
}

df = df_raw[list(cols_to_keep.keys())].rename(columns=cols_to_keep)
df = df.iloc[1:].reset_index(drop=True)
df["zip_code"] = df["area_name"].str.extract(r"(\d{5})")

numeric_cols = ["total_population","white_alone","black_alone","asian_alone","hispanic_or_latino","total_housing_units","moe_total_housing_units"]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df = df.dropna(subset=["total_population"])
df["pct_white"] = df["white_alone"] / df["total_population"]
df["pct_black"] = df["black_alone"] / df["total_population"]
df["pct_asian"] = df["asian_alone"] / df["total_population"]
df["pct_hispanic"] = df["hispanic_or_latino"] / df["total_population"]

df_clean = df[[
    "zip_code",
    "total_population",
    "white_alone",
    "black_alone",
    "asian_alone",
    "hispanic_or_latino",
    "pct_white",
    "pct_black",
    "pct_asian",
    "pct_hispanic",
    "total_housing_units",
    "moe_total_housing_units"
]]

Path("data/01-clean").mkdir(parents=True, exist_ok=True)
df_clean.to_csv("data/01-clean/acs_zip_demographics_clean.csv", index=False)

print(df_clean.shape)
```
